```python
from scipy import stats
import numpy as np

s = df['Bugs_Reported']
n_used, n_missing = s.notna().sum(), s.isna().sum()
W, p = stats.shapiro(s)
z = (s - s.mean()) / s.std()
outliers = s[z.abs() > 3]

print(f'mean={s.mean():.2f}  median={s.median()}  std={s.std():.2f}  max={s.max()}')
print(f'skew={stats.skew(s):.2f} (>0 => правый хвост)   kurtosis={stats.kurtosis(s):.2f} (>0 => тяжёлый хвост)')
print(f'Shapiro-Wilk: W={W:.4f}, p={p:.1e} -> ' + ('НЕ нормальное' if p < 0.05 else 'нормальное'))
print(f'Выбросов |z|>3: {len(outliers)} шт., значения от {outliers.min()} до {outliers.max()} '
      f'(макс. z = {z.abs().max():.2f})')

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].hist(s, bins=range(0, int(s.max())+2), color='steelblue', edgecolor='white')
ax[0].axvline(s.mean(),   color='red',   ls='--', lw=2, label=f'mean={s.mean():.2f}')
ax[0].axvline(s.median(), color='green', ls='-',  lw=2, label=f'median={s.median():.0f}')
thr = s.mean() + 3*s.std()
ax[0].axvspan(thr, s.max()+1, color='orange', alpha=0.25, label='зона выбросов |z|>3')
ax[0].set_title('Bugs_Reported - гистограмма'); ax[0].set_xlabel('bugs'); ax[0].legend()
# боксплот: наглядно видно хвост выбросов справа
ax[1].boxplot(s, vert=False)
ax[1].set_title('Bugs_Reported - boxplot (точки справа = выбросы)'); ax[1].set_xlabel('bugs')
plt.tight_layout(); plt.show()
```

## 5. z-score: кандидаты в аномалии

## 7. Сместить выборку

## 8. Сравните две группы (t-тест)


```python
from scipy import stats
import numpy as np
g0 = df[df.Task_Success==0]['AI_Usage_Hours']
g1 = df[df.Task_Success==1]['AI_Usage_Hours']
t, p = stats.ttest_ind(g0, g1, equal_var=False)   # Welch
sp = np.sqrt(((len(g0)-1)*g0.var() + (len(g1)-1)*g1.var()) / (len(g0)+len(g1)-2))
d = (g1.mean()-g0.mean()) / sp
size = 'БОЛЬШОЙ' if abs(d)>=0.8 else 'СРЕДНИЙ' if abs(d)>=0.5 else 'МАЛЕНЬКИЙ'
print(f'group 0 (провал): n={len(g0)} mean={g0.mean():.3f}')
print(f'group 1 (успех):  n={len(g1)} mean={g1.mean():.3f}')
print(f'Welch t={t:.3f}, p={p:.2e}  -> {"значимо" if p<0.05 else "не значимо"}')
print(f"Cohen's d = {d:.3f}  -> эффект {size}")
print('\nНезависимость: 100 разработчиков x ~10 строк каждый -> строки НЕ независимы,')
print('это повторные измерения, истинный df меньше, p занижен (преувеличивает значимость).')
print('строк на разработчика:', df['Developer_ID'].value_counts().agg(['min','mean','max']).to_dict())
```

## Особый вызов: охота на призрак (p-hacking)
Сгенерируем **20 колонок чистого шума** (`np.random.normal` — заведомо не связаны ни с чем), переберём все пары с реальными колонками и найдём самую «впечатляющую» корреляцию и самое «значимое» различие. Затем покажем ловушку: **вот цифра, которая убеждает — и вот почему она ничего не значит.**

**Догадка:** при ~180 проверках на уровне α=0.05 чисто случайно вылезет около **9 «значимых»** ложных результатов, и какой-нибудь шум покажет p < 0.01, хотя связи нет.

```python
import numpy as np
from scipy import stats
np.random.seed(42)

real = ['Hours_Coding','AI_Usage_Hours','Lines_of_Code','Commits','Bugs_Reported',
        'Sleep_Hours','Distractions','Cognitive_Load','Stress_Level']
noise = {f'noise_{i}': np.random.normal(size=len(df)) for i in range(20)}

# 1) перебор корреляций шум x реальные
res = []
for n, col in noise.items():
    for rc in real:
        r, p = stats.pearsonr(col, df[rc])
        res.append((abs(r), r, p, n, rc))
res.sort(reverse=True)
_, r, p, n, rc = res[0]
sig = sum(1 for x in res if x[2] < 0.05)
print(f'Проверено корреляций: {len(res)} (20 шумовых x 9 реальных)')
print(f'САМАЯ впечатляющая ложная связь: {n} ~ {rc}: r={r:+.3f}, p={p:.4f}')
print(f'"Значимых" p<0.05: {sig}  (ожидается случайно: {0.05*len(res):.0f})')
print(f'Порог Бонферрони = 0.05/{len(res)} = {0.05/len(res):.5f} -> '
      f'призрак выживает? {p < 0.05/len(res)}')

# 2) групповые различия на СЛУЧАЙНЫХ разрезах
best = None
for i in range(20):
    grp = np.random.randint(0, 2, len(df))
    for rc in real:
        t, pp = stats.ttest_ind(df[rc][grp==0], df[rc][grp==1], equal_var=False)
        if best is None or pp < best[1]:
            best = (rc, pp, i, t)
print(f'\nСамое "значимое" различие на СЛУЧАЙНОМ разрезе: {best[0]} '
      f'(разрез #{best[2]}): t={best[3]:.2f}, p={best[1]:.4f}')

# визуализация: лучшая ложная связь выглядит убедительно
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.scatter(noise[n], df[rc], alpha=0.4, color='purple')
m,c0 = np.polyfit(noise[n], df[rc], 1)
xs = np.array([noise[n].min(), noise[n].max()])
plt.plot(xs, m*xs+c0, color='red', lw=2, label=f'r={r:+.3f}, p={p:.3f}')
plt.xlabel(n + ' (чистый шум)'); plt.ylabel(rc)
plt.title('Призрак: убедительная связь из ничего'); plt.legend(); plt.show()
```

### Ловушка, названная вслух
**Вот цифра, которая убеждает:** `noise_7 ~ Sleep_Hours`, **r ≈ 0.10, p = 0.0019** — в отрыве это выглядит как «значимое открытие» (p < 0.01!).

**Вот почему она ничего не значит:** `noise_7` — это буквально `np.random.normal`, в нём нет никакой информации. Мы сделали **180 слепых попыток**, и при α=0.05 примерно **9 ложных срабатываний гарантированы случайно** — мы получили 17. Ни один призрак **не выживает** после поправки Бонферрони (порог 0.00028). Это и есть **p-hacking**: перебери достаточно вариантов — и случай сам выдаст «результат».

**Связь с переобучением и R²=0.99:** это тот же механизм. Когда модель получает много признаков (а тем более шумовых), она находит случайные совпадения в обучающих данных и показывает почти идеальный R² — который **рассыпается на новых данных**. Высокий R² на обучении убеждает ровно так же, как наш p=0.0019, и врёт ровно по той же причине.

**ВЫВОД:** самая «значимая» связь из 180 проверок (r=0.10, p=0.0019) — чистый шум; при стольких попытках ~9 ложных хитов неизбежны.
**ПРОВЕРКА:** noise_7 — это np.random.normal без всякой связи; ни один результат не проходит порог Бонферрони 0.00028.

> **Самое ценное открытие дня — не то, что сказали данные, а то, где они чуть нас не обманули:** число с p<0.01, за которым стоит абсолютная пустота.
