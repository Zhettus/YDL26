# Wider Hyperparameter Search — Side Experiment (`wider_tuning_test.ipynb`)

**Self-contained side-experiment.** It does **not** modify `hack2.ipynb`, and it will only ever
write a *new* file `submission_wider.csv` — never the existing `submission.csv`.

## What we're testing

Part 10 tuned the stack's members over **narrow** grids and concluded that tuning **ties within
noise**. A separate notebook used a **wider, more aggressive** search and reported a slightly
higher score. So the fair question is: *does a wider search over **our existing three members** —
CatBoost, XGBoost, SVC-RBF — find real gains Part 10 missed?*

- We **only re-tune the models already in the champion stack**. We do **not** add any new model
  (e.g. HistGradientBoosting). This isolates "did we under-tune?" from "did they add a model we
  chose not to."
- Everything is scored and selected on **macro-F1** (`scoring="f1_macro"`) — the competition
  metric — **never accuracy**. Selecting on accuracy would optimize the wrong thing for an
  imbalance-sensitive macro metric.

## Facts carried from the project

- `train.csv` (9,000; fallback `tran.csv`), `final-check.csv` (5,000). Target `sleep_stage`
  (4 classes). `id` = row index → drop from features, keep for submission.
- Features = 21 + `eog_burst_index_missing` flag (1 where NaN, computed from raw NaNs *before*
  imputation).
- **Outer evaluation** = `StratifiedKFold(5, shuffle=True, random_state=42)` (the whole
  project's folds). **Inner search** = a separate `StratifiedKFold(5, shuffle=True,
  random_state=42)` — we use **5-fold inner** (not the other notebook's 3-fold) to keep the search
  itself low-noise.
- **Champion benchmark:** `Stacking([CatBoost, XGBoost, SVC-RBF(prob=True)], final=LogReg,
  internal cv=5, predict_proba)` → **CV macro-F1 0.8308 ± 0.0052, Kaggle 0.8328**.

## Two honesty caveats (state them up front)

1. `RandomizedSearchCV.best_score_` is **optimistic** — it's the best over many configs *peeked at
   on all inner folds*, so it flatters the winner.
2. Because the inner and outer folds share `random_state=42`, even our **outer re-evaluation of a
   tuned member** isn't fully held out from the search. The cleanest signals are therefore the
   tuned **stack's** outer CV (stacking adds its own internal CV) and, ultimately, **Kaggle**. We
   still apply the strict **> 1 outer-std** rule, but we read it with this caveat in mind.

## 1. Setup

Load data, build the same feature matrix and folds as the project, and define `eval_full` (one
leakage-free outer-fold loop → mean ± std **and** out-of-fold predictions for stage-2 F1).

```python
# Installs if needed:
# %pip install catboost xgboost
import os, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV, GridSearchCV
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from catboost import CatBoostClassifier
from xgboost import XGBClassifier

RANDOM_STATE = 42
TARGET = "sleep_stage"
FLAG_COL = "eog_burst_index_missing"

def _load_first_existing(candidates):
    for name in candidates:
        if os.path.exists(name):
            return pd.read_csv(name), name
    raise FileNotFoundError(f"None found: {candidates}")

train_df, train_file = _load_first_existing(["train.csv", "tran.csv"])
test_df,  test_file  = _load_first_existing(["final-check.csv"])
ID_COLS = [c for c in ["id"] if c in train_df.columns]
base_features = [c for c in train_df.columns if c not in ID_COLS + [TARGET]]
X = train_df[base_features].copy()
y = train_df[TARGET].copy()
X[FLAG_COL] = X["eog_burst_index"].isna().astype(int)
model_features = base_features + [FLAG_COL]

outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)  # project folds
inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)  # 5-fold search
classes = sorted(y.unique())
print(f"train={train_file} {X.shape}, test={test_file} {test_df.shape}, features={len(model_features)}")

def make_pipeline(model):
    return Pipeline([("impute", SimpleImputer(strategy="median")),
                     ("scale",  StandardScaler()),
                     ("model",  model)])

def eval_full(factory, label="model", verbose=True):
    Xv = X[model_features]
    oof = np.empty(len(Xv), dtype=int)
    fold_f1, t0 = [], time.time()
    for tr, va in outer_cv.split(Xv, y):
        est = factory(); est.fit(Xv.iloc[tr], y.iloc[tr])
        oof[va] = np.asarray(est.predict(Xv.iloc[va])).ravel().astype(int)
        fold_f1.append(f1_score(y.iloc[va], oof[va], average="macro"))
    mean, std = float(np.mean(fold_f1)), float(np.std(fold_f1))
    s2 = f1_score(y, oof, average=None, labels=classes)[classes.index(2)]
    if verbose:
        print(f"{label:30s} macro-F1 = {mean:.4f} +/- {std:.4f}   stage2-F1 = {s2:.4f}  "
              f"({time.time()-t0:.0f}s)")
    return {"label": label, "mean": mean, "std": std, "stage2_f1": s2, "oof": oof}

def strip(p): return {k.split('__',1)[1]: v for k,v in p.items()}

BENCH_MEAN, BENCH_STD = 0.8308, 0.0052
print(f"Champion benchmark: {BENCH_MEAN} +/- {BENCH_STD}")
```

## 2. Tune CatBoost — wide RandomizedSearchCV (`scoring="f1_macro"`, `n_iter=25`)

Wide space: `depth`, `learning_rate`, `l2_leaf_reg`, `iterations`, `border_count`. We set
`thread_count=1` on CatBoost and let `RandomizedSearchCV(n_jobs=-1)` parallelize across the 25×5
fits — this avoids CPU oversubscription (model threads × search threads) that would *slow* things.

```python
cat_space = {
    "model__depth":         [4, 6, 8, 10],
    "model__learning_rate": [0.02, 0.03, 0.05, 0.08, 0.1],
    "model__l2_leaf_reg":   [1, 3, 5, 7, 9],
    "model__iterations":    [400, 600, 800, 1200],
    "model__border_count":  [32, 64, 128],
}
cat_search = RandomizedSearchCV(
    make_pipeline(CatBoostClassifier(random_state=RANDOM_STATE, verbose=0,
                                     allow_writing_files=False, thread_count=1)),
    cat_space, n_iter=25, scoring="f1_macro", cv=inner_cv,
    random_state=RANDOM_STATE, n_jobs=-1, refit=True)
t0 = time.time(); cat_search.fit(X[model_features], y)
cat_best = strip(cat_search.best_params_)
print(f"CatBoost search {time.time()-t0:.0f}s | best (optimistic) f1_macro={cat_search.best_score_:.4f}")
print("Best params:", cat_best)

def f_cat_default(): return make_pipeline(CatBoostClassifier(iterations=600, learning_rate=0.05,
                                          depth=6, random_state=RANDOM_STATE, verbose=0,
                                          allow_writing_files=False))
def f_cat_tuned():   return make_pipeline(CatBoostClassifier(random_state=RANDOM_STATE, verbose=0,
                                          allow_writing_files=False, **cat_best))
cat_def  = eval_full(f_cat_default, "CatBoost default")
cat_tune = eval_full(f_cat_tuned,   "CatBoost tuned (wide)")
print(f"CatBoost delta macro-F1: {cat_tune['mean']-cat_def['mean']:+.4f}")
```

## 3. Tune XGBoost — wide RandomizedSearchCV (`scoring="f1_macro"`, `n_iter=25`)

Wide space over `n_estimators`, `max_depth`, `learning_rate`, `subsample`, `colsample_bytree`,
`reg_lambda`, `min_child_weight`. XGBoost set to `n_jobs=1` so the search's `n_jobs=-1` does the
parallelism cleanly.

```python
xgb_space = {
    "model__n_estimators":     [300, 500, 800, 1200],
    "model__max_depth":        [3, 4, 6, 8],
    "model__learning_rate":    [0.02, 0.05, 0.08, 0.1],
    "model__subsample":        [0.7, 0.85, 1.0],
    "model__colsample_bytree": [0.7, 0.85, 1.0],
    "model__reg_lambda":       [1, 3, 5, 7],
    "model__min_child_weight": [1, 3, 5],
}
xgb_search = RandomizedSearchCV(
    make_pipeline(XGBClassifier(random_state=RANDOM_STATE, eval_metric="mlogloss",
                                tree_method="hist", n_jobs=1)),
    xgb_space, n_iter=25, scoring="f1_macro", cv=inner_cv,
    random_state=RANDOM_STATE, n_jobs=-1, refit=True)
t0 = time.time(); xgb_search.fit(X[model_features], y)
xgb_best = strip(xgb_search.best_params_)
print(f"XGBoost search {time.time()-t0:.0f}s | best (optimistic) f1_macro={xgb_search.best_score_:.4f}")
print("Best params:", xgb_best)

def f_xgb_default(): return make_pipeline(XGBClassifier(n_estimators=400, learning_rate=0.05,
                                          max_depth=6, subsample=0.9, colsample_bytree=0.9,
                                          random_state=RANDOM_STATE, eval_metric="mlogloss",
                                          n_jobs=-1, tree_method="hist"))
def f_xgb_tuned():   return make_pipeline(XGBClassifier(random_state=RANDOM_STATE,
                                          eval_metric="mlogloss", tree_method="hist", n_jobs=-1,
                                          **xgb_best))
xgb_def  = eval_full(f_xgb_default, "XGBoost default")
xgb_tune = eval_full(f_xgb_tuned,   "XGBoost tuned (wide)")
print(f"XGBoost delta macro-F1: {xgb_tune['mean']-xgb_def['mean']:+.4f}")
```

## 4. Tune SVC-RBF — wider GridSearchCV (`scoring="f1_macro"`)

Part 10 found SVC's defaults (`C=1`, `gamma="scale"`) optimal over a *narrower* range. Here we
widen `C` up to 20 and add more `gamma` values to confirm or update that. We search with
`probability=False` (faster; macro-F1 from `predict` doesn't need calibrated probabilities), then
rebuild the winner with `probability=True` for the stack.

```python
svc_grid = {"model__C": [0.5, 1, 2, 5, 10, 20],
            "model__gamma": ["scale", 0.01, 0.03, 0.05, 0.1, 0.3]}
svc_search = GridSearchCV(
    make_pipeline(SVC(kernel="rbf", probability=False, random_state=RANDOM_STATE)),
    svc_grid, scoring="f1_macro", cv=inner_cv, n_jobs=-1, refit=True)
t0 = time.time(); svc_search.fit(X[model_features], y)
svc_best = strip(svc_search.best_params_)
print(f"SVC grid {time.time()-t0:.0f}s | best (optimistic) f1_macro={svc_search.best_score_:.4f}")
print("Best params:", svc_best)

def f_svc_default(): return make_pipeline(SVC(kernel="rbf", probability=False,
                                              random_state=RANDOM_STATE))
def f_svc_tuned():   return make_pipeline(SVC(kernel="rbf", probability=False,
                                              random_state=RANDOM_STATE, **svc_best))
svc_def  = eval_full(f_svc_default, "SVC default")
svc_tune = eval_full(f_svc_tuned,   "SVC tuned (wide)")
print(f"SVC delta macro-F1: {svc_tune['mean']-svc_def['mean']:+.4f}")
print("(If best params == C=1, gamma=scale, the wider range CONFIRMS Part 10.)")
```

## 5. Rebuild the stack with the tuned members

Same structure as the champion (final = LogReg, internal `cv=5`, `predict_proba`), but each base
now uses its wide-search best params. The SVC base is rebuilt with `probability=True` (stacking
needs probabilities). We also re-evaluate the **default** stack in this same run for an exact
same-fold comparison.

> Slow cells: each stack fits SVC many times via the internal CV. Let them run.

```python
def f_svc_tuned_proba(): return make_pipeline(SVC(kernel="rbf", probability=True,
                                                  random_state=RANDOM_STATE, **svc_best))
def f_svc_def_proba():   return make_pipeline(SVC(kernel="rbf", probability=True,
                                                  random_state=RANDOM_STATE))

def build_stack(cat_f, xgb_f, svc_f):
    return StackingClassifier(
        estimators=[("catboost", cat_f()), ("xgboost", xgb_f()), ("svc_rbf", svc_f())],
        final_estimator=LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
        cv=5, stack_method="predict_proba", n_jobs=-1)

stack_def  = eval_full(lambda: build_stack(f_cat_default, f_xgb_default, f_svc_def_proba),
                       "Stack DEFAULT (champion)")
stack_tune = eval_full(lambda: build_stack(f_cat_tuned, f_xgb_tuned, f_svc_tuned_proba),
                       "Stack TUNED (wide)")
```

## 6. Scoreboard + honesty check

The decision rule (programmatic): the tuned stack is a **real win** only if it beats the default
stack by **more than one outer-fold std (~0.005)** on macro-F1. Inside that band = **TIE** → keep
the champion. We also check whether **stage-2 F1** moved.

```python
rows = [("CatBoost default", cat_def), ("CatBoost tuned", cat_tune),
        ("XGBoost default", xgb_def),  ("XGBoost tuned", xgb_tune),
        ("SVC default", svc_def),      ("SVC tuned", svc_tune),
        ("Stack DEFAULT", stack_def),  ("Stack TUNED", stack_tune)]
board = (pd.DataFrame([{"config": n, "macro_f1": r["mean"], "std": r["std"],
                        "stage2_f1": r["stage2_f1"]} for n, r in rows])
           .sort_values("macro_f1", ascending=False).reset_index(drop=True))
print(board.round(4).to_string(index=False))

dStack = stack_tune["mean"] - stack_def["mean"]
WIN = dStack > stack_def["std"]
print(f"\nTuned stack {stack_tune['mean']:.4f} vs default {stack_def['mean']:.4f} -> "
      f"delta {dStack:+.4f} (need > +{stack_def['std']:.4f})")
print(f"stage-2 F1: default {stack_def['stage2_f1']:.3f} -> tuned {stack_tune['stage2_f1']:.3f} "
      f"(delta {stack_tune['stage2_f1']-stack_def['stage2_f1']:+.3f})")
print("=> REAL WIN beyond noise." if WIN else "=> TIE within noise -> KEEP the champion.")
```

```python
# Confusion comparison (leakage-free OOF): did anything change on the stage-2/3 boundary?
def norm_cm(oof):
    cm = confusion_matrix(y, oof, labels=classes); return cm / cm.sum(axis=1, keepdims=True)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, run, tag in [(axes[0], stack_def, "Stack DEFAULT"), (axes[1], stack_tune, "Stack TUNED")]:
    sns.heatmap(norm_cm(run["oof"]), annot=True, fmt=".2f", cmap="Blues", vmin=0, vmax=1,
                xticklabels=classes, yticklabels=classes, ax=ax)
    ax.set_title(tag); ax.set_xlabel("predicted"); ax.set_ylabel("true")
plt.tight_layout(); plt.show()
r2, c3 = classes.index(2), classes.index(3)
print(f"stage2->stage3 confusion: default {norm_cm(stack_def['oof'])[r2,c3]*100:.1f}% -> "
      f"tuned {norm_cm(stack_tune['oof'])[r2,c3]*100:.1f}%")
```

## 7. Verdict + (only if it wins) build `submission_wider.csv`

If the tuned stack beat the champion **beyond noise**, we retrain it on **all** training data,
predict `final-check.csv`, and write **`submission_wider.csv`** (never `submission.csv`) with full
sanity checks. If it merely tied, we conclude the champion stands.

```python
if WIN:
    print("Tuned stack won beyond noise -> building submission_wider.csv\n")
    # Build identically-prepared test frame.
    test_ids = test_df["id"].copy()
    X_test = test_df.drop(columns=["id"]).copy()
    X_test[FLAG_COL] = X_test["eog_burst_index"].isna().astype(int)
    X_test = X_test[model_features]
    assert list(X_test.columns) == list(model_features), "column mismatch!"

    final = build_stack(f_cat_tuned, f_xgb_tuned, f_svc_tuned_proba)
    final.fit(X[model_features], y)                       # fit on ALL training rows
    pred = np.asarray(final.predict(X_test)).ravel().astype(int)

    sample_cols = ["id", TARGET]
    for s in ["sample_submission.csv", "submission.csv"]:
        if os.path.exists(s):
            sample_cols = list(pd.read_csv(s, nrows=1).columns); break
    sub = pd.DataFrame({sample_cols[0]: test_ids.values, sample_cols[1]: pred})[sample_cols]

    ok = True
    c1 = len(sub) == len(test_df);                         ok &= c1
    print(f"[{'PASS' if c1 else 'FAIL'}] 5000 rows: {len(sub)}")
    c2 = sub[sample_cols[1]].isna().sum() == 0;            ok &= c2
    print(f"[{'PASS' if c2 else 'FAIL'}] no NaN predictions")
    c3 = sub[sample_cols[1]].dropna().apply(lambda v: float(v).is_integer()).all() \
         and set(sub[sample_cols[1]].unique()).issubset({0,1,2,3});  ok &= c3
    print(f"[{'PASS' if c3 else 'FAIL'}] integer labels in {{0,1,2,3}}")
    dist = sub[sample_cols[1]].value_counts(normalize=True).sort_index()
    c4 = (dist >= 0.05).all();                             ok &= c4
    print(f"[{'PASS' if c4 else 'FAIL'}] every class >= 5%:\n{(dist*100).round(1)}")
    c5 = list(sub.columns) == sample_cols;                 ok &= c5
    print(f"[{'PASS' if c5 else 'FAIL'}] columns == {sample_cols}")

    if ok:
        sub.to_csv("submission_wider.csv", index=False)
        print("\nAll checks passed -> wrote submission_wider.csv (submission.csv untouched).")
        print(sub.head().to_string(index=False))
    else:
        print("\nSanity checks FAILED -> NOT writing submission_wider.csv.")
else:
    print("TIE within noise -> NOT building a submission. The champion stands "
          "(CV 0.8308, Kaggle 0.8328). submission.csv left untouched.")
```

## 8. Takeaway

Read the scoreboard and verdict above:

- **If the wide-tuned stack beat the champion beyond one outer-fold std** (and especially if
  stage-2 moved): wider tuning *did* find gains the narrow Part-10 search missed → consider
  `submission_wider.csv` for upload (Kaggle is the final arbiter, given the inner/outer-fold
  optimism caveat).
- **If it tied (the expected outcome):** tuning our existing members **harder and wider** still
  lands inside the noise — confirming Part 10 over a much broader range. That means the other
  notebook's small edge most plausibly came from a **model we deliberately didn't add
  (HistGradientBoosting)** plus ordinary submission/leaderboard noise — **not** from under-tuning
  CatBoost / XGBoost / SVC. The champion stands.

A tie here is a *strong* result: it says our 0.8308 stack is near the practical ceiling for these
three members, and effort is better spent (if anywhere) on genuinely new geometry than on more
tuning.

> Leakage discipline held: every transformer fit inside its CV; searches used a proper inner CV;
> the final fit (only on a win) trains on all-train and merely applies to test; `submission.csv`
> is never overwritten — only `submission_wider.csv` is ever written.
