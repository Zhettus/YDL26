```python
x = np.linspace(0, 2*np.pi, 200)
plt.plot(x, np.sin(x), label="sinX")
plt.legend() 
plt.savefig("sin", dpi=300) #save the figure and dpi make quality better
plt.show()
#Q: 
#what linspace does, what arguments do
#what is meshgrid, what arguments do
#what is cmap
```

```python
x = np.linspace(0, np.pi*4, 100)
plt.plot(x, np.sin(x), color="green", linestyle="-", marker="1")
plt.grid(True, alpha=0.3) #alpha is transparency of the grid
plt.show() #do everything necessary before show, after show its deleted
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.show()
```
